```python
list1 = list()
mylabels = list()
#for gen in movies.Genre.unique():
for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre == gen].BudgetMillion)
    mylabels.append(gen)
#print(list1) 

sns.set_style('darkgrid')
fig,ax = plt.subplots()
fig.set_size_inches(11.7, 8.27) #size of A4 paper
h = plt.hist(list1,bins = 15, stacked = True, rwidth = 1,label = mylabels) # color = ["Red","Blue","Green",..] you can add seven colors according to number of legends
plt.title("Movie Budget Distribution", fontsize = 35, \
         color = "Red" , fontname = "Console")
plt.ylabel("Number of Movies", Fontsize = 25, color = "Blue")
plt.xlabel("Budget", Fontsize = 25, color = "Green")
plt.yticks(fontsize = 20)
plt.xticks(fontsize = 20)


plt.legend(frameon = True,fancybox = True, shadow = True, framealpha = 0.5, prop = {'size':20}) # for plotting legends
plt.show()
```
